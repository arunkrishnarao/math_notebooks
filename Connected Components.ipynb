```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread("Resources/art8.png", 0)
img = cv2.resize(img, (256, 256))
img = img.astype('uint64')

threshold = 40
img[img <= threshold] = 0 
img[img > threshold] = 1
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img, cmap = 'gray')
```

```python
visited = []
pixels = []

def four_connected_component(i, j, set):
    global pixels, img, visited
    (p1, q1) = (i, j-1)
    (p2, q2) = (i, j+1)
    (p3, q3) = (i-1, j)
    (p4, q4) = (i+1, j)
    
    if q1 != -1 and (img[p1][q1] in set) and ((p1, q1) not in visited):
        pixels.append((p1, q1))
        visited.append((p1, q1))
        four_connected_component(p1, q1, set)

    if q2 < h and (img[p2][q2] in set) and ((p2, q2) not in visited):
        pixels.append((p2, q2))
        visited.append((p2, q2))
        four_connected_component(p2, q2, set)

    if p3 != -1 and (img[p3][q3] in set) and ((p3, q3) not in visited):
        pixels.append((p3, q3))
        visited.append((p3, q3))
        four_connected_component(p3, q3, set)

    if p4 < w and (img[p4][q4] in set) and ((p4, q4) not in visited):
        pixels.append((p4, q4))
        visited.append((p4, q4))
        four_connected_component(p4, q4, set)
    return pixels
```

```python
v = {1}
print(v)
four_components = []

(w, h) = img.shape 
for i in range(w):
    for j in range(h):
        if img[i][j] in v and ((i, j) not in visited):
            pixels = []
            pixels = four_connected_component(i, j, v)
            if pixels:
                visited.append((i, j))
                four_components.append(pixels)
```

```python
import matplotlib.pyplot as plt
import random
image_connected = np.zeros_like(img)

fig = plt.figure()
for component in four_components:
    color = random.randint(1, 255)
    for point in component:
        (i, j) = point
        image_connected[i][j] = color
plt.imshow(image_connected, cmap = 'gray')
```
