## Bresenham Line Drawing Algorithm

Given endpoints $(x_1, y_1)$ and $(x_2, y_2)$, find all the intermediate pixels between them:


```python
import pandas as pd
table = []

print("Enter the Starting point: ")
x1 = int(input("x1: "))
y1 = int(input("y1: "))

print("\nEnter the Ending point: ")
x2 = int(input("x2: "))
y2 = int(input("y2: "))

dx = x2 - x1
dy = y2 - y1
print("\nSolution")

print("dx = x2 - x1 = ", dx)
print("dy = y2 - y1 = ", dy)

# Compute Decision factor
p = 2 * dy - dx
i = 0
print(f"\np{i} = 2*dy - dx = ", p)
print(f"p{i} = 2 * {dy} - {dx} = ", p, "\n")

table.append([i, p, x1, y1])
 
for i in range(1, dx+1):
    if p < 0: # Negetive Decision Factor
        print(f"=> p{i} = ", p ,", p is negetive")
        x1 += 1

        print(f"=> Increment x1, x1 = {x1} and y1 = {y1}\n")
        print(f"p{i+1} = p{i} + 2*dy\n=> p{i+1} = {p} + {2*dy}" )
        table.append([i, p, x1, y1])
        p = p + 2*dy

    else: # Positive Decision Factor
        print(f"=> p{i} = ", p ,", p is possitive")
        x1 += 1
        y1 += 1

        print(f"=> Increment both x1 and y1, x1 = {x1} and y1 = {y1}\n")
        print(f"p{i+1} = p{i} + 2*dy - 2*dx\n=> p{i+1} = {p} + {2*dy} - {2*dx}" )
        table.append([i, p, x1, y1])
        p = p + 2*dy - 2*dx


df = pd.DataFrame(table)
df.columns =['i', 'decision factor', 'x-value', 'y-value']
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 5), dpi=100)
plt.xlim(-1, x2+1)
plt.ylim(-1, y2+1)

plt.plot(df['x-value'], df['y-value'], marker="o", markersize=5, markerfacecolor="red")
plt.grid()
```
